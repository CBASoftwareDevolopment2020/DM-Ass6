```python
operators = ("+", "-", "<", "<=", "==", ">=", ">", "and", "or", "not")

class Node:
    
    def __init__(self, val=None, func=None, block=False):
        self.left = None
        self.right = None
        self.val = val
        self.block = block
        self.mid = []
        if block:
            self.val = "BLOCK"
        
    def __repr__(self):
        return str(self.val)

    
    
def compile(s):
    root = Node(block=True)
    words = s.split()
    print(s)
    
    i=0
    
    while(i < len(words)):
        word = words[i]
        
        if word == "def":
            new_node = Node(word, "def")
            new_node.left = Node(words[i+1][:-1], "val")
            new_node.right = Node(words[i+2], "type")
            
            root.mid.append(new_node)
            i=i+2# Skip next 2 words
            
        
        if word == "let":
            new_node = Node(word, "let")
            new_node.left = Node(words[i+1], "val")
            
            # let y = y + 10
            if i+5 < len(words) and words[i+4] in operators:
                ope_node = Node(words[i+4], "ope")
                ope_node.left = Node(words[i+3], "val")
                ope_node.right = Node(words[i+5], "val")
                
                new_node.right = ope_node
                i=i+5
            # let y = 10
            else:
                new_node.right = Node(words[i+3], "val")
                i=i+3
                
            root.mid.append(new_node)
        
        
        
        if word in ("if", "while"):
            new_node = Node(word, word)
            root.mid.append(new_node)
            
            if words[i+3] == ")":
                # if ( TRUE ) { ...... }
                # if ( x ) { ...... }
                ope_node = Node("==", "ope")
                ope_node.left = Node("TRUE", "val")
                ope_node.right = Node(words[i+2], "val")
                new_node.mid.append(ope_node)
                i=i+3
                
            elif words[i+1] == "NOT":
                # if ( NOT  x ) { ...... }
                ope_node = Node(words[i+2], "ope")
                ope_node.left = Node(words[i+3], "val")
                new_node.mid.append(ope_node)
                i=i+3
            
            elif words[i+3] in operators:
                # if ( x == 5) { ...... }
                ope_node = Node(words[i+3], "ope")
                ope_node.left = Node(words[i+2], "val")
                ope_node.right = Node(words[i+4], "val")
                new_node.mid.append(ope_node)
                i=i+6
            
            for j in range(i, len(words)):
                if words[j] == "}":
                    
                    new_block = compile(" ".join(words[i+1:j]))
                    new_node.mid.append(new_block)
                    i = j+1
                    break
                        
            if  word == "if" and j+1 < len(words) and words[j+1] == "else":
                for k in range(j+1, len(words)):
                    if words[k] == "}":
                        new_block = compile(" ".join(words[j+3:k]))
                        new_node.mid.append(new_block)
                        i = k+1
                        break            
        else:
            i=i+1

            
    return root


    

#n = compile("if ( a == b ) { let y = 10 } else { let z == 7 } let o = 420")
n = compile("while ( i <= 10 ) { let i = i + 1 }")

print(n.mid)
# if statement has "==, BLOCK, BLOCK"
print(n.mid[0].mid[1].mid[0].right.right)
```

```python
code = """
def x int
def y int
def z bool

let x = 5
let y = 14

let x = x * 3

let z = x > y

let z = z and true
"""

n = compile(code)

def parse(data):
    v = {}
    
    def is_int(x):
        try: 
            int(x)
            return type(x) != bool
        except ValueError:
            return False
        
    def is_bool(x):
        return type(x) == bool or x == "false" or x == "true"
    
    def fix_type(x, v):
        if is_int(x):
            x = int(x)
        elif is_bool(x):
            x = x == "true"
        else:
            x = v[x]
        return x
    
    # operators = ("+", "*", "-", "<", "<=", "==", ">=", ">", "AND", "OR", "NOT")
    def evalu(ope, a, b, v):

        a = fix_type(a, v)
        b = fix_type(b, v)
        
        if (type(a) != type(b)):
            raise Exception(f"can't compare {a} and {b}")
            
        if(is_int(a)):
            a = int(a)
            b = int(b)
            if ope=="+":
                return a+b
            elif ope=="*":
                return a*b
            elif ope=="-":
                return a-b
            elif ope=="<":
                return a<b
            elif ope=="<=":
                return a<=b
            elif ope=="==":
                return a==b
            elif ope==">=":
                return a>=b
            elif ope==">":
                return a>b
        elif(is_bool(a)):
            if ope=="==":
                return a==b
            elif ope=="and":
                print("AND",a,b)
                return a and b
            elif ope=="or":
                return a or b
            elif ope=="not":
                return not a
        else:
            raise Exception(f"weird comparison between '{a}' and '{b}'")
        
        return True
    
    def dig(v, data):
        if data[0] == "block":
            print("is block")
            for s in data[1:]:
                if s[0] == "def":
                    if s[2][1] == "int": # set to 0 if int
                        v[s[1][1]] = 0
                    elif s[2][1] == "bool": # set to false if bool
                        v[s[1][1]] = False
                    else:
                        return False
                
                if s[0] == "let":
                    if s[1][1] not in v: # var not defined
                        raise Exception(str(s[1][1]) + " var not defined")
                    
                    if s[2][0] == "ope":
                        res = evalu(s[2][1], s[2][2][1], s[2][3][1], v)
                    else:
                        res = fix_type(s[2][1], v)
                        
                    if type(v[s[1][1]]) != type(res): # check if var is int or bool
                        raise Exception(res + " is wrong type")
                    else:
                        v[s[1][1]] = res
                        
                    
    dig(v,data)
    return v
    
parse(n)
```
